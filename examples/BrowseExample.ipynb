# MAAP Browse Example

### Load the MAAP python library

```python
import warnings
from maap.maap import MAAP
warnings.filterwarnings('ignore')
maap = MAAP('/maap-py/maap.cfg')
```

### Display a UAVSAR granule

```python
granule = maap.searchGranule(granule_ur='uavsar_AfriSAR_v1_SLC-lopenp_14043_16015_001_160308_L090.vrt')[0]
maap.show(granule)
```

### Display an LVIS granule

```python
granule = maap.searchGranule(granule_ur='ILVIS2_GA2016_0220_R1611_038024')[0]
maap.show(granule)
```
